```python
# replace bbox
# replace dates
# replace client id / client secret
# create loops for rolling window
# create loops for extracting QA

import openeo
from pathlib import Path
from shapely.geometry import box

openeo_backend = "https://openeofed.dataspace.copernicus.eu"
openeo_backend_client_id = ""
openeo_backend_client_secret = ""
start_date = "2020-01-01"
end_date = "2020-03-01"
minx, miny, maxx, maxy = 22.045441, 39.019184, 22.754059, 39.498742
bbox = box(minx, miny, maxx, maxy)

connection = openeo.connect(openeo_backend).authenticate_oidc_client_credentials(
    client_id=openeo_backend_client_id, client_secret=openeo_backend_client_secret
)

connection.describe_collection("CGLS_SSM_V1_EUROPE")
collection = connection.load_collection(
    collection_id="CGLS_SSM_V1_EUROPE",
    spatial_extent=bbox,
    temporal_extent=[start_date, end_date],
)

composite = collection.reduce_dimension(
    dimension="t", reducer="median"
)

output_file = Path(Path(".").resolve(), "composite.tif")
title = "sample_CGLS_SSM_V1_EUROPE_median"
composite.execute_batch(output_file, out_format="GTiff", title=title)
```
